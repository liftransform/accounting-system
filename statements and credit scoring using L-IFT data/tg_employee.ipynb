```python
import pandas as pd
import numpy as np
```

```python
employee_data = pd.read_excel(r'lift_data\Employee Report - Mon Feb 13 2023.xlsx', sheet_name='Employee payments')
```

```python
employee_data.head()
```

```python
employee_data.reset_index(inplace=True)
```

```python
employee_data.columns
```

```python
employee_cols = [
    'index', 'Respondent ID', 'Employee payment employee data ID',
    'Employee payment employee name', 'Employee payment amount',
    'Employee payment date', 
]
```

```python
AMOUNT = 'Employee payment amount'
EMP_ID = 'Employee payment employee data ID'

debit_acc_col = np.full(len(employee_data), 'salary expense')
credit_acc_col = np.full(len(employee_data), 'cash')
debit_amts = employee_data[AMOUNT]
credit_amts = employee_data[AMOUNT]

transactions = pd.DataFrame(
    {
        'debit acc': debit_acc_col,
        'credit acc': credit_acc_col,
        'debit amt': debit_amts,
        'credit amt': credit_amts
    }
)
transactions.reset_index(inplace=True)

employee_transactions = pd.merge(employee_data[employee_cols], transactions, how='outer', on='index')
employee_transactions
```

```python
employee_transactions.to_csv('lift_data_transactions/employee.csv')
```
