```python
def fill_gaps(data: pd.DataFrame, cols:list[str]):
    new_data:pd.DataFrame = data.copy()
    for col in cols:
        new_series = []
        previous = None
        for d in data[col]:
            if pd.isna(d):
                new_series.append(previous)
            else:
                previous = d
                new_series.append(d)
        new_data.drop(columns=col, inplace=True)
        new_data[col] = new_series
    return new_data

asset_filtered2 = fill_gaps(asset_filtered, ['Respondent ID', 'Asset data ID', 'Asset possesion type'])
asset_filtered2
```

```python
asset_filtered2['Asset type'].fillna('', inplace=True)
```

```python
asset_filtered2 = asset_filtered2.reset_index()
asset_filtered2.head()
```

```python
def check_condition(condition:dict, row: pd.Series) -> bool:
    for key, value in condition.items():
        if type(value) == list:
            found_one = False
            for v in value:
                if v in row[key]:
                    found_one = True
                    break
            if found_one:
                continue
            return False
        elif str(value) in str(row[key]) :
            continue

        return False
    return True

def record_default(row, id_col_name='index'):
    return [row[id_col_name], '', '', '', '']

def record_transaction(cond_map:list[tuple], frame: pd.DataFrame, id_col_name = 'id') -> pd.DataFrame:
    transactions = []
    for row in frame.iterrows():
        recorded = False
        for cond, func in cond_map:
            if check_condition(cond, row[1]):
                transactions.append(func(row[1]))
                recorded = True
                break
        if not recorded:
            transactions.append(record_default(row[1], id_col_name))

    trans_frame = pd.DataFrame(transactions, columns=[id_col_name, 'debit acc', 'debit amt', 'credit acc', 'credit amt'])
    return pd.merge(frame, trans_frame, 'outer', id_col_name)

def record_func_creator(debit_acc, credit_acc, id_col_name, amount_col):
    def func(row: pd.Series):
        return [row[id_col_name], debit_acc, row[amount_col], credit_acc, row[amount_col]]

    return func
```

```python
AMOUNT_COL = 'Asset amount'
ASSET_ID = 'index'

ASSET_TYPE = 'Asset type'
OWNERSHIP = 'Asset possesion type'

LAND = 'land'
BUILDING = 'building'
LIVESTOCK = 'live stock'
PURCHASES = 'purchases'
EQUIPMENT = 'equipment'
EQUITY = 'equity'

condition_func_map = [
    
    (
        {
            ASSET_TYPE: ['Inventory'],
        },
        record_default
    ),
    (
        {
            ASSET_TYPE: ['Livestock'],
            OWNERSHIP: ['Own']
        },
        record_func_creator(LIVESTOCK, EQUITY, ASSET_ID, AMOUNT_COL)
    ),
    (
        {
            ASSET_TYPE: ['Building'],
            OWNERSHIP: ['Own']
        },
        record_func_creator(BUILDING, EQUITY, ASSET_ID, AMOUNT_COL)
    ),
    (
        {
            ASSET_TYPE: ['Land'],
            OWNERSHIP: ['Own']
        },
        record_func_creator(LAND, EQUITY, ASSET_ID, AMOUNT_COL)
    ),
    (
        {
            ASSET_TYPE: [
                'Computers', 'Electrical appliances', 'Furniture',
                'Machinery', 'Office equipment', 'Other', 'Telephones',
                'Vehicles', 'Tools & implements', ''
            ],
            OWNERSHIP: ['Own']
        },
        record_func_creator(EQUIPMENT, EQUITY, ASSET_ID, AMOUNT_COL)
    ),

]
```

```python
asset_transactions = record_transaction(condition_func_map, asset_filtered2, ASSET_ID)
asset_transactions
```
